### 1. Start the Kafka Simulator
This will spin up a background Python process that generates emergency calls and pushes them to our local Kafka broker (`kafka:9092`).

*(Note: Any previously running simulator is killed first to prevent orphaned background processes).* 

```python
import subprocess
import time
import os

# Ensure no old simulator processes are running in the background
os.system("pkill -f emergency_call_simulator.py")
time.sleep(1)

# Start simulator in the background
print("Starting Emergency Call Simulator...")
simulator_process = subprocess.Popen(
    ["python", "/home/jovyan/ingestion/kafka/producers/emergency_call_simulator.py"],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True
)

time.sleep(3)
print("Simulator is running in the background!")
```

### 2. Start the Spark Streaming Job
Now we'll submit the Spark streaming processor (`streaming_processor.py`) to the Spark Master.
This job continuously reads from Kafka, performs transformations, and streams the data into our ClickHouse `Gold` layer.

*Note: The `--packages` flag is required for Kafka integration. Since we've now persisted the Ivy cache, it will only resolve dependencies without re-downloading them over the internet!*

```python
print("Submitting Spark Streaming Job...")
!spark-submit --master spark://spark-master:7077 --packages org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.0 /opt/spark/jobs/streaming_processor.py
```
